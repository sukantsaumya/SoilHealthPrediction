```python
# Load and print the training ground truth data
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/archive/train_data/train_data/train_gt.csv')
print(df)

# Load and print the wavelengths
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/archive/train_data/train_data/wavelengths.csv')
print(df)

# Check the image size
import numpy as np

# Load the .npz file
data = np.load('/content/drive/MyDrive/archive/test_data/test_data/10.npz')

# Assuming one of the keys contains the image data, let's find its size
for key in data.files:
    array = data[key]
    if len(array.shape) == 3:
        print(f"Size of image data stored in '{key}': {array.shape[0]} x {array.shape[1]} x {array.shape[2]}")
    elif len(array.shape) == 2:
        print(f"Size of image data stored in '{key}': {array.shape[0]} x {array.shape[1]}")

# Visualize test data
import numpy as np
import matplotlib.pyplot as plt

# Load the .npz file
data = np.load('/content/drive/MyDrive/archive/test_data/test_data/10.npz')

# Check the keys in the .npz file
print("Keys in .npz file:", data.files)

# Assuming one of the keys contains the spectral data, let's visualize them
for key in data.files:
    array = data[key]
    if len(array.shape) >= 1:  # Ensure we're dealing with at least 1D data
        num_curves = array.shape[0]  # Number of spectral curves
        plt.figure(figsize=(8, 6))
        for i in range(min(num_curves, 10)):  # Limit to first 10 curves if there are many
            plt.plot(array[i])
        plt.title(key)
        plt.xlabel('Wavelength')
        plt.ylabel('Intensity')
        plt.grid(True)
        plt.show()

# Import necessary libraries for file operations
from os import listdir
from os.path import isfile, join
import glob
import matplotlib.pyplot as plt
import skimage as sk

# Load test data into a dictionary
import numpy as np
import glob
from PIL import Image

# Read all .npz files in the directory
dataset = {}
for npz_file in sorted(glob.glob("/content/drive/MyDrive/archive/test_data/test_data/*.npz")):
    # Extract the filename (without extension) as the key
    filename = npz_file.split('/')[-1].split('.')[0]
    # Load the .npz file
    data = np.load(npz_file)
    # Get the first key if 'data' doesn't exist
    if 'data' in data:
        image_data = data['data']
    else:
        image_data = data[list(data.keys())[0]]
    dataset[filename] = image_data

print(dataset.keys())

# Visualize the first 10 wavelength bands for an image
if "1" in dataset:
    img = dataset["1"]
    # Check if img is 3D (has wavelength bands)
    if len(img.shape) == 3:
        plt.figure(figsize=(20, 20))
        for i in range(min(10, img.shape[2])):
            plt.subplot(2, 5, i+1)
            plt.imshow(img[:, :, i])
            plt.colorbar()
            plt.title("Wavelength " + str(i))
    else:
        print("Image does not have wavelength bands (shape:", img.shape, ")")

# Define desired width and height for image resizing
desired_width = 64
desired_height = 64

# List files in the test data directory
from os import listdir
from os.path import isfile, join
import glob
import matplotlib.pyplot as plt
import skimage as sk

# iterate in directory and extract all images names
mypath='/content/drive/MyDrive/archive/test_data/test_data'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))] # get all images names
onlyfiles.sort() # sort by lowest-highest wavelength
print(mypath)

# CNN model attempt 1
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Define paths
train_data_dir = '/content/drive/MyDrive/archive/train_data/train_data'
test_data_dir = '/content/drive/MyDrive/archive/test_data/test_data'
train_gt_path = '/content/drive/MyDrive/archive/train_data/train_data/train_gt.csv'
wavelengths_path = '/content/drive/MyDrive/archive/train_data/train_data/wavelengths.csv'

# Load train data
train_data_files = [os.path.join(train_data_dir, f) for f in os.listdir(train_data_dir) if f.endswith('.npz')]
train_data_list = []

for f in train_data_files:
    data = np.load(f)
    # Check if 'data' key exists
    if 'data' in data:
        train_data_list.append(data['data'])
    else:
        # Use the first available key
        train_data_list.append(data[list(data.keys())[0]])

# Ensure all data has the same shape or reshape as needed
if train_data_list:
    # Check dimensions
    shapes = [data.shape for data in train_data_list]
    print("Shapes of training data:", shapes)

    # Check if all data has 3 dimensions (height, width, channels)
    if all(len(shape) == 3 for shape in shapes):
        # Find the maximum height and width
        max_height = max(shape[0] for shape in shapes)
        max_width = max(shape[1] for shape in shapes)

        # Pad the arrays to the same size
        padded_data_list = []
        for data in train_data_list:
            pad_height = max_height - data.shape[0]
            pad_width = max_width - data.shape[1]
            padded_data = np.pad(data, ((0, pad_height), (0, pad_width), (0, 0)), mode='constant')
            padded_data_list.append(padded_data)

        # Concatenate arrays
        train_data = np.stack(padded_data_list)
    else:
        print("Error: Data dimensions are not consistent")
        # Attempt to reshape or flatten data
        train_data = np.array([data.flatten() for data in train_data_list])
else:
    print("No training data found")
    train_data = np.array([])

# Load ground truth labels
train_gt = pd.read_csv(train_gt_path)

# Load wavelengths
wavelengths = pd.read_csv(wavelengths_path)

# Check if we have data to proceed
if len(train_data) > 0:
    # Extract features and labels
    X_train = train_data

    # Ensure train_gt has the expected columns
    if all(col in train_gt.columns for col in ['P', 'K', 'MG', 'PH']):
        y_train = train_gt[['P', 'K', 'MG', 'PH']].values
    else:
        print("Warning: Expected columns not found in train_gt. Available columns:", train_gt.columns)
        # Use first 4 columns as labels
        y_train = train_gt.iloc[:, 1:5].values  # Skip first column (likely filenames)

    # Check if X_train and y_train have compatible shapes
    if len(X_train) != len(y_train):
        print(f"Error: X_train has {len(X_train)} samples but y_train has {len(y_train)} samples")
        # Use the minimum number of samples
        min_samples = min(len(X_train), len(y_train))
        X_train = X_train[:min_samples]
        y_train = y_train[:min_samples]

    # Preprocess the data
    scaler = StandardScaler()

    # Reshape if needed for StandardScaler
    orig_shape = X_train.shape
    X_train_reshaped = X_train.reshape(X_train.shape[0], -1)
    X_train_scaled = scaler.fit_transform(X_train_reshaped)
    X_train_scaled = X_train_scaled.reshape(orig_shape)

    # Define the CNN architecture (adjust input_shape as needed)
    input_shape = X_train_scaled.shape[1:]

    model = tf.keras.models.Sequential([
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape, padding='same'),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(4)  # 4 output neurons for P, K, MG, and PH values
    ])

    # Compile the model
    model.compile(optimizer='adam', loss='mse')  # Mean Squared Error loss for regression

    # Train the model
    model.fit(X_train_scaled, y_train, epochs=10, batch_size=32, validation_split=0.2)

    # Load test data
    test_data_files = [os.path.join(test_data_dir, f) for f in os.listdir(test_data_dir) if f.endswith('.npz')]
    test_data_list = []

    for f in test_data_files:
        data = np.load(f)
        # Check if 'data' key exists
        if 'data' in data:
            test_data_list.append(data['data'])
        else:
            # Use the first available key
            test_data_list.append(data[list(data.keys())[0]])

    # Ensure all test data has the same shape
    if test_data_list:
        # Pad the arrays to the same size as the training data
        padded_test_data_list = []
        for data in test_data_list:
            # Ensure test data has the same shape as input_shape
            if data.shape[1:] != input_shape[1:]:  # Compare only width, height, channels
                pad_height = input_shape[0] - data.shape[0]
                pad_width = input_shape[1] - data.shape[1]
                if pad_height >= 0 and pad_width >= 0:
                    padded_data = np.pad(data, ((0, pad_height), (0, pad_width), (0, 0)), mode='constant')
                    padded_test_data_list.append(padded_data)
                else:
                    # If test data is larger, crop it
                    padded_data = data[:input_shape[0], :input_shape[1], :]
                    padded_test_data_list.append(padded_data)
            else:
                padded_test_data_list.append(data)

        # Concatenate arrays
        test_data = np.stack(padded_test_data_list)

        # Preprocess the test data
        test_data_reshaped = test_data.reshape(test_data.shape[0], -1)
        test_data_scaled = scaler.transform(test_data_reshaped)
        X_test_scaled = test_data_scaled.reshape(test_data.shape)

        # Make predictions
        predictions = model.predict(X_test_scaled)

        # Print predictions for the first sample in the test data
        print("Predicted P value:", predictions[0, 0])
        print("Predicted K value:", predictions[0, 1])
        print("Predicted MG value:", predictions[0, 2])
        print("Predicted PH value:", predictions[0, 3])
    else:
        print("No test data found")
else:
    print("No training data to train the model")

# CNN model attempt 2 (fixed) - Using proper train/test split
import glob
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Load training dataset
train_data = {}
for npz_file in sorted(glob.glob("/content/drive/MyDrive/archive/train_data/train_data/*.npz")):
    filename = npz_file.split('/')[-1].split('.')[0]
    data = np.load(npz_file)
    # Check if 'data' key exists
    if 'data' in data:
        image_data = data['data']
    else:
        # Use the first available key
        image_data = data[list(data.keys())[0]]
    train_data[filename] = image_data

# Load testing dataset
test_data = {}
for npz_file in sorted(glob.glob("/content/drive/MyDrive/archive/test_data/test_data/*.npz")):
    filename = npz_file.split('/')[-1].split('.')[0]
    data = np.load(npz_file)
    # Check if 'data' key exists
    if 'data' in data:
        image_data = data['data']
    else:
        # Use the first available key
        image_data = data[list(data.keys())[0]]
    test_data[filename] = image_data

# Load wavelengths
wavelengths = pd.read_csv("/content/drive/MyDrive/archive/train_data/train_data/wavelengths.csv")

# Check if we have data to proceed
if train_data and wavelengths is not None:
    # Get the number of wavelengths
    num_wavelengths = wavelengths.shape[0]

    # Preprocess the data
    scaler = StandardScaler()

    # Prepare training data
    X_train_list = []
    for image in train_data.values():
        # Handle different image shapes
        if len(image.shape) == 3 and image.shape[2] == num_wavelengths:
            # Image shape is (height, width, wavelengths)
            # Resize to a common shape if needed
            resized_image = np.resize(image, (desired_height, desired_width, num_wavelengths))
            X_train_list.append(resized_image)
        elif len(image.shape) == 2:
            # Image shape is (pixels, wavelengths)
            # Reshape to (height, width, wavelengths)
            resized_image = np.resize(image, (desired_height, desired_width, 1))
            X_train_list.append(resized_image)
        else:
            print(f"Skipping image with unexpected shape: {image.shape}")

    if X_train_list:
        X_train = np.stack(X_train_list)

        # Prepare test data
        X_test_list = []
        for image in test_data.values():
            # Handle different image shapes
            if len(image.shape) == 3 and image.shape[2] == num_wavelengths:
                # Image shape is (height, width, wavelengths)
                resized_image = np.resize(image, (desired_height, desired_width, num_wavelengths))
                X_test_list.append(resized_image)
            elif len(image.shape) == 2:
                # Image shape is (pixels, wavelengths)
                resized_image = np.resize(image, (desired_height, desired_width, 1))
                X_test_list.append(resized_image)
            else:
                print(f"Skipping test image with unexpected shape: {image.shape}")

        if X_test_list:
            X_test = np.stack(X_test_list)

            # Load ground truth labels
            train_labels_df = pd.read_csv("/content/drive/MyDrive/archive/train_data/train_data/train_gt.csv")

            # Check if 'filename' column exists
            if 'filename' in train_labels_df.columns:
                train_labels_df.set_index('filename', inplace=True)

                # Extract labels for each training sample
                y_train_list = []
                for filename in train_data.keys():
                    if filename in train_labels_df.index:
                        if all(col in train_labels_df.columns for col in ['P', 'K', 'MG', 'PH']):
                            y_train_list.append(train_labels_df.loc[filename][['P', 'K', 'MG', 'PH']].values)
                        else:
                            # Use columns 0-3 as a fallback
                            y_train_list.append(train_labels_df.loc[filename].iloc[:4].values)
                    else:
                        print(f"Warning: No label found for filename {filename}")

                if y_train_list:
                    y_train = np.array(y_train_list)

                    # Check if X_train and y_train have the same number of samples
                    if len(X_train) != len(y_train):
                        print(f"Warning: X_train has {len(X_train)} samples but y_train has {len(y_train)} samples")
                        # Use the minimum number of samples
                        min_samples = min(len(X_train), len(y_train))
                        X_train = X_train[:min_samples]
                        y_train = y_train[:min_samples]

                    # Split training data into training and validation sets
                    X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

                    # Define the CNN architecture
                    input_shape = X_train_split.shape[1:]

                    model = Sequential([
                        Conv2D(32, (3, 3), activation='relu', input_shape=input_shape, padding='same'),
                        MaxPooling2D((2, 2)),
                        Conv2D(64, (3, 3), activation='relu', padding='same'),
                        MaxPooling2D((2, 2)),
                        Conv2D(64, (3, 3), activation='relu', padding='same'),
                        Flatten(),
                        Dense(64, activation='relu'),
                        Dense(4)  # 4 output neurons for P, K, MG, and PH values
                    ])

                    # Compile the model
                    model.compile(optimizer='adam', loss='mse')  # Mean Squared Error loss for regression

                    # Train the model
                    history = model.fit(X_train_split, y_train_split, epochs=10, batch_size=32, validation_data=(X_val, y_val))

                    # Evaluate the model
                    loss = model.evaluate(X_val, y_val)
                    print("Validation Loss:", loss)

                    # Make predictions on test data
                    predictions = model.predict(X_test)
                    print("Predictions for test data:", predictions[:5])  # Show first 5 predictions
                else:
                    print("No labels available for training data")
            else:
                print("Error: 'filename' column not found in train_gt.csv")
        else:
            print("No test data available after preprocessing")
    else:
        print("No training data available after preprocessing")
else:
    print("Missing required data: training data or wavelengths")

# Classification model attempt - Fixed version
import numpy as np
import pandas as pd
import glob
import os
from PIL import Image
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks

# Load training data
train_dataset = {}
for npz_file in sorted(glob.glob("/content/drive/MyDrive/archive/train_data/train_data/*.npz")):
    filename = npz_file.split('/')[-1].split('.')[0]
    data = np.load(npz_file)
    # Check if 'data' key exists
    if 'data' in data:
        image_data = data['data']
    else:
        # Use the first available key
        image_data = data[list(data.keys())[0]]
    train_dataset[filename] = image_data

# Load testing data
test_dataset = {}
for npz_file in sorted(glob.glob("/content/drive/MyDrive/archive/test_data/test_data/*.npz")):
    filename = npz_file.split('/')[-1].split('.')[0]
    data = np.load(npz_file)
    # Check if 'data' key exists
    if 'data' in data:
        image_data = data['data']
    else:
        # Use the first available key
        image_data = data[list(data.keys())[0]]
    test_dataset[filename] = image_data

# Load ground truth labels
train_labels_path = "/content/drive/MyDrive/archive/train_data/train_data/train_gt.csv"
if os.path.exists(train_labels_path):
    train_labels_df = pd.read_csv(train_labels_path)

    # Check if we need to create class labels from P, K, MG, PH values
    if 'class' not in train_labels_df.columns and all(col in train_labels_df.columns for col in ['P', 'K', 'MG', 'PH']):
        # Create a simple classification by binning the values
        # Example: create class based on the highest value
        cols = ['P', 'K', 'MG', 'PH']
        train_labels_df['class'] = train_labels_df[cols].idxmax(axis=1)
        # Convert string classes to integers
        class_mapping = {cls: i for i, cls in enumerate(train_labels_df['class'].unique())}
        train_labels_df['class_id'] = train_labels_df['class'].map(class_mapping)

    # Handle case where columns are not named
    elif train_labels_df.shape[1] >= 5:  # Assuming filename + 4 values (P, K, MG, PH)
        # Rename columns
        if train_labels_df.columns[0] == 0:  # Unnamed columns
            new_columns = ['filename', 'P', 'K', 'MG', 'PH']
            train_labels_df.columns = new_columns[:train_labels_df.shape[1]]
            # Create class based on highest value
            cols = train_labels_df.columns[1:5]  # P, K, MG, PH
            train_labels_df['class'] = train_labels_df[cols].idxmax(axis=1)
            class_mapping = {cls: i for i, cls in enumerate(train_labels_df['class'].unique())}
            train_labels_df['class_id'] = train_labels_df['class'].map(class_mapping)

    # Ensure filename is used as index
    if 'filename' in train_labels_df.columns:
        train_labels_df.set_index('filename', inplace=True)

    # Load wavelengths
    wavelengths_path = "/content/drive/MyDrive/archive/train_data/train_data/wavelengths.csv"
    if os.path.exists(wavelengths_path):
        wavelengths = pd.read_csv(wavelengths_path)
        num_wavelengths = len(wavelengths)

        # Preprocess data
        X_train = []
        y_train = []

        for key, value in train_dataset.items():
            if key in train_labels_df.index:
                # Process image data
                if len(value.shape) == 3:
                    # Image has 3 dimensions (height, width, channels)
                    # Resize to fixed dimensions
                    resized_image = np.resize(value, (desired_height, desired_width, value.shape[2]))
                    X_train.append(resized_image)
                elif len(value.shape) == 2:
                    # Image has 2 dimensions - expand to 3D
                    resized_image = np.resize(value, (desired_height, desired_width, 1))
                    X_train.append(resized_image)
                else:
                    print(f"Skipping image with unexpected shape: {value.shape}")
                    continue

                # Get class label
                if 'class_id' in train_labels_df.columns:
                    y_train.append(train_labels_df.loc[key]['class_id'])
                elif 'class' in train_labels_df.columns:
                    y_train.append(train_labels_df.loc[key]['class'])
                else:
                    # Use first column as class
                    y_train.append(train_labels_df.loc[key].iloc[0])
            else:
                print(f"Warning: No label found for filename {key}")
                continue

        if X_train and y_train:
            X_train = np.array(X_train)
            y_train = np.array(y_train)

            # Process test data
            X_test = []
            for value in test_dataset.values():
                if len(value.shape) == 3:
                    # Image has 3 dimensions (height, width, channels)
                    resized_image = np.resize(value, (desired_height, desired_width, value.shape[2]))
                    X_test.append(resized_image)
                elif len(value.shape) == 2:
                    # Image has 2 dimensions - expand to 3D
                    resized_image = np.resize(value, (desired_height, desired_width, 1))
                    X_test.append(resized_image)
                else:
                    print(f"Skipping test image with unexpected shape: {value.shape}")

            if X_test:
                X_test = np.array(X_test)

                # Check input shape consistency
                input_shape = (desired_height, desired_width, X_train.shape[3] if len(X_train.shape) > 3 else 1)

                # Determine number of classes
                num_classes = len(np.unique(y_train))

                # Define CNN model
                model = models.Sequential([
                    layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape, padding='same'),
                    layers.MaxPooling2D((2, 2)),
                    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
                    layers.MaxPooling2D((2, 2)),
                    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
                    layers.Flatten(),
                    layers.Dense(64, activation='relu'),
                    layers.Dense(num_classes, activation='softmax')
                ])

                model.compile(optimizer='adam',
                            loss='sparse_categorical_crossentropy',
                            metrics=['accuracy'])

                # Train the model
                history = model.fit(X_train, y_train, epochs=10, validation_split=0.2)

                # Predict
                predictions = model.predict(X_test)
                predicted_classes = np.argmax(predictions, axis=1)

                # Print accuracy
                print("Predictions for test data:", predicted_classes)

                # If we have a class mapping, convert back to original class names
                if 'class_id' in train_labels_df.columns and 'class' in train_labels_df.columns:
                    reverse_mapping = {v: k for k, v in class_mapping.items()}
                    predicted_class_names = [reverse_mapping.get(cls_id, "Unknown") for cls_id in predicted_classes]
                    print("Predicted class names:", predicted_class_names)
            else:
                print("No test data available after preprocessing")
        else:
            print("No training data available after preprocessing")
    else:
        print(f"Wavelengths file not found: {wavelengths_path}")
else:
    print(f"Training labels file not found: {train_labels_path}")
```
